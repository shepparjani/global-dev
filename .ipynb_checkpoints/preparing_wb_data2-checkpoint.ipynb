```python
mortality_fileloc = "resources/mortality_rate_under5.csv"
co2total_fileloc = "resources/co2_emissions_kt.csv"
co2pc_fileloc = "resources/CO2_mt_percapita.csv"
electricity_fileloc = "resources/electricity_access.csv"
electr_consum_fileloc = "resources/electricity_cons.csv"
gdp_fileloc = "resources/gdp.csv"
```

```python
co2pc_df = pd.read_csv(co2pc_fileloc)
co2pc_df = co2pc_df.rename(columns={"2018":"CO2percap_2018"})
selected_columns = co2pc_df[['Country Name', 'Country Code', 'CO2percap_2018']]
co2pc2018_df = selected_columns.copy()
co2pc2018_df.to_csv("resources/co2pc2018.csv", index=False)
co2pc2018_df.head(5)
```

```python
electricitydf = pd.read_csv(electricity_fileloc)
electricitydf = electricitydf.rename(columns={"2018":"Electricity_2018"})
selected_columns = electricitydf[['Country Name', 'Country Code', 'Electricity_2018']]
electricity2018_df = selected_columns.copy()
electricity2018_df.to_csv("resources/electricity2018.csv", index=False)
electricity2018_df.head(5)
```

```python
mortalitydf = pd.read_csv(mortality_fileloc)
mortalitydf = mortalitydf.rename(columns={"2018":"Mortality_2018"})
selected_columns = mortalitydf[['Country Name', 'Country Code', 'Mortality_2018']]
mortality2018_df = selected_columns.copy()
mortality2018_df.to_csv("resources/mortality2018.csv", index=False)
mortality2018_df.head(5)
```

```python
elect_consume_df = pd.read_csv(electr_consum_fileloc)
elect_consume_df = elect_consume_df.rename(columns={"2014":"Electr_Cons_2014"})
selected_columns = elect_consume_df[['Country Name', 'Country Code', 'Electr_Cons_2014']]
electr_consume_2014_df = selected_columns.copy()
electr_consume_2014_df.to_csv("resources/electrconsume2014.csv", index=False)
electr_consume_2014_df.head(5)
```

```python
gdp_df = pd.read_csv(gdp_fileloc)
gdp_df = gdp_df.rename(columns={"2020":"gdppc_2020"})
selected_columns = gdp_df[['Country Name', 'Country Code', 'gdppc_2020']]
gdppc2020_df = selected_columns.copy()
gdppc2020_df.to_csv("resources/gdp2020.csv", index=False)
gdppc2020_df.head(5)
```

```python
gdppc2020_df['gdppc_2020'].unique()
```

```python
import pandas as pd
from functools import reduce

# compile the list of dataframes you want to merge
dfwrldbnk_merged = [mortality2018_df, electricity2018_df, co2pc2018_df, electr_consume_2014_df, gdppc2020_df]
dfwrldbnk_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Country Code'],
                                            how='outer'), dfwrldbnk_merged)
# need to look at the above line of code again. 
selected_columns = dfwrldbnk_merged[['Country Name', 'Country Code', 'Mortality_2018', 'Electricity_2018', 'CO2percap_2018','Electr_Cons_2014', 'gdppc_2020']]
dfwrldbnk_merged = selected_columns.copy()
dfwrldbnk_merged.to_csv("resources/combinedwrldbnkdata.csv", index=False)
dfwrldbnk_merged.head(5)
```
